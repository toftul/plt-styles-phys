```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('phys-plots.mplstyle')
```

```python
plt.style.use('seaborn-colorblind')
```

```python
x = np.linspace(0, stop=4*np.pi, num=300)
y = np.sin(x)**4

plt.plot(x, y, label='$\sin^4(x)$')
plt.plot(x, y**2, label='$\sin^8(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.legend(loc=1)
plt.show()
```
