# Custom styles for better plots

Can be loaded as simplte as
- Computer modern style

```python
plt.style.use('https://raw.githubusercontent.com/toftul/plt-styles-phys/main/phys-plots-cm.mplstyle')
```

- Improved sans style

```python
plt.style.use('https://raw.githubusercontent.com/toftul/plt-styles-phys/main/phys-plots-sans.mplstyle')
```

Or you are free to download these style files, change and use as
```python
plt.style.use('<FILENAME>.mplstyle')
```

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.linspace(0, stop=2*np.pi, num=300)
y1 = np.sin(x)**100
y2 = np.sin(x)**10
````

## Default settings 

````python
plt.rcdefaults()

plt.figure(figsize=(4, 5))
plt.title('Default')
plt.plot(x, y1, label='$\sin^{100}(x)$')
plt.plot(x, y2, label='$\sin^{50}(x)$')
plt.xlabel('Important value, $x$')
plt.ylabel('Arbitrary units, $y$')

plt.legend(loc=1)
plt.tight_layout()
plt.savefig('figs/fig3.png')
plt.show()
````

## Computer moder font (default $\LaTeX$ font)

````python
plt.rcdefaults()
plt.style.use('phys-plots-cm.mplstyle')

plt.figure(figsize=(4, 5))
plt.title('Computer modern')
plt.plot(x, y1, label='$\sin^{100}(x)$')
plt.plot(x, y2, label='$\sin^{50}(x)$')
plt.xlabel('Important value, $x$')
plt.ylabel('Arbitrary units, $y$')

plt.legend(loc=1)
plt.tight_layout()
plt.savefig('figs/fig1.png')
plt.show()
````

## Good sans font (ideal for presentations)

````python
plt.rcdefaults()
plt.style.use('phys-plots-sans.mplstyle')

plt.figure(figsize=(4, 5))
plt.title('Improved sans')
plt.plot(x, y1, label='$\sin^{100}(x)$')
plt.plot(x, y2, label='$\sin^{50}(x)$')
plt.xlabel('Important value, $x$')
plt.ylabel('Arbitrary units, $y$')

plt.legend(loc=1)
plt.tight_layout()
plt.savefig('figs/fig2.png')
plt.show()
````
